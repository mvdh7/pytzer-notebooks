```julia
using Interact, Plots
include("../PitzerModel.jl")

ions = [:Na,:K,:Mg,:Ca,:Cl]
tempK = 298.15

ioncharges = Model.Properties.ioncharges(ions)

amin = 3
amult = 3
@manipulate for mNa = 0:0.01:1, mK = 0:0.01:1, mMg = 0:0.01:1, mCa = 0:0.01:1
    
    mCl = mNa + mK + 2*mMg + 2*mCa
    
    mols = [mNa,mK,mMg,mCa,mCl]
    
     = Model.(mols,tempK,ioncharges)
    
    for (i,mol) in enumerate(mols)
    
        i == 1 ? scatterx = scatter : scatterx = scatter!
        
        scatterx([mol,],[[i],] , markersize=mol*[i] * amult + amin, 
            markerstrokealpha=0, grid=true, legend=true, label=ions[i])
        
    end  

    xlims!((0,8.1))
    ylims!((0,3))
end
```
