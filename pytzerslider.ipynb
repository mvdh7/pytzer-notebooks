```julia
using Interact, Plots
include("../Pitzer.jl/PitzerModel.jl")

ions = [:Na,:K,:Mg,:Ca,:Cl,:H,:OH]
tempK = 298.15

ioncharges = Model.Properties.ioncharges(ions)

amin = 3
amult = 4
@manipulate for mNa  = 0:0.0094:0.94,
                mK   = 0:0.0002:0.02,
                mMg  = 0:0.001:0.1,
                mCa  = 0:0.0002:0.02,
                mCl  = 0:0.011:1.1,
                mAll = 0:0.02:2
    
    mX = mNa + mK + 2*mMg + 2*mCa - mCl
    
    mX > 0 ? (mOH = mX; mH = 0) : (mH = -mX; mOH = 0)
    
    mols = [mNa,mK,mMg,mCa,mCl,mH,mOH] * mAll
    
    γ = Model.γ(mols,tempK,ioncharges)
    
    for (i,mol) in enumerate(mols)
    
        i == 1 ? scatterx = scatter : scatterx = scatter!
        
        scatterx([sqrt(mol),],[γ[i],] , markersize=mol*γ[i] * amult + amin, 
            markerstrokealpha=0, grid=true, legend=true, label=ions[i])
        
    end  

    xlims!((0,1.6))
    ylims!((0,1.5))
    
    xlabel!("sqrt(m)")
    ylabel!("acf")
    
end
```
